# Analisis Data Idol K-pop: Karakteristik dan Distribusi
Proyek ini bertujuan untuk menganalisis dataset idol K-pop guna memahami karakteristik fisik, asal negara, distribusi gender, serta ukuran grup idol. Analisis ini disusun sebagai bagian dari portofolio data analyst pemula.


## 1. Data Understanding
Pada tahap ini, kita akan memuat data dan melihat struktur awalnya.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('kpop idols.csv')

print('Struktur Data:')
df.info()
print('\n5 Baris Pertama:')
df.head()
```

## 2. Data Cleaning
Pembersihan data dilakukan untuk memastikan hasil analisis akurat. Langkah-langkahnya meliputi:
- Mengganti nama kolom agar lebih mudah dibaca.
- Menangani nilai 0 pada tinggi dan berat badan (dianggap missing value).
- Mengubah format tanggal lahir menjadi tipe datetime.
- Menangani nilai yang hilang pada kolom Group.


```python
# Renaming columns
new_columns = {
    'Stage Name Stage Name': 'Stage Name',
    'Full Name Full Name': 'Full Name',
    'Korean Name Korean Name': 'Korean Name',
    'K. Stage Name K. Stage Name': 'K. Stage Name',
    'Date of Birth Date of Birth': 'DOB',
    'Group Group': 'Group',
    'Country Country': 'Country',
    'Height Height': 'Height',
    'Weight Weight': 'Weight',
    'Birthplace Birthplace': 'Birthplace',
    'Gender Gender': 'Gender'
}
df.rename(columns=new_columns, inplace=True)

# Handling 0 values in physical stats
df['Height'] = df['Height'].replace(0, np.nan)
df['Weight'] = df['Weight'].replace(0, np.nan)

# Date processing
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['Birth Year'] = df['DOB'].dt.year

print('Cleaning Selesai!')
```

## 3. Exploratory Data Analysis (EDA)


### Topik 1: Karakteristik Fisik Idol Berdasarkan Gender
Pertanyaan: Bagaimana perbedaan rata-rata tinggi badan dan berat badan idol K-pop berdasarkan gender?


```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Gender', y='Height', data=df, palette='pastel')
plt.title('Distribusi Tinggi Badan per Gender')
plt.ylabel('Tinggi (cm)')

plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='Weight', data=df, palette='pastel')
plt.title('Distribusi Berat Badan per Gender')
plt.ylabel('Berat (kg)')

plt.tight_layout()
plt.show()

print('Rata-rata Tinggi dan Berat:')
print(df.groupby('Gender')[['Height', 'Weight']].mean())
```

**Insight:**
- Idol laki-laki memiliki rata-rata tinggi badan yang lebih tinggi dibanding idol perempuan (~177 cm vs ~164 cm).
- Variasi berat badan juga menunjukkan pola serupa, di mana idol laki-laki cenderung lebih berat.


### Topik 2: Persebaran Negara Asal
Pertanyaan: Negara asal mana yang paling banyak muncul?


```python
top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Negara Asal Idol K-pop')
plt.xlabel('Jumlah Idol')
plt.ylabel('Negara')
plt.show()
```

**Insight:**
- Mayoritas besar idol berasal dari South Korea.
- Negara-negara seperti Japan, China, dan USA merupakan kontributor utama berikutnya, menunjukkan ekspansi internasional industri K-pop.


### Topik 3: Distribusi Gender
Pertanyaan: Bagaimana komposisi idol laki-laki dan perempuan?


```python
plt.figure(figsize=(7, 7))
counts = df['Gender'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['skyblue', 'pink'], startangle=90)
plt.title('Persentase Gender Idol K-pop')
plt.show()
```

**Insight:**
- Distribusi gender dalam dataset ini sangat berimbang antara pria (M) dan wanita (F).


### Topik 4: Grup dengan Anggota Terbanyak
Pertanyaan: Grup mana yang paling besar skalanya dalam dataset?


```python
top_groups = df.groupby('Group')['Stage Name'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_groups.values, y=top_groups.index, palette='rocket')
plt.title('10 Grup dengan Anggota Terbanyak dalam Dataset')
plt.xlabel('Jumlah Anggota')
plt.ylabel('Nama Grup')
plt.show()
```

**Insight:**
- NCT memiliki jumlah anggota terbanyak yang terdaftar, sesuai dengan konsep mereka sebagai grup global dengan sistem sub-unit yang fleksibel.


### Topik 5: Karakteristik Tahun Lahir
Pertanyaan: Pada periode tahun berapa idol paling banyak lahir?


```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Birth Year'].dropna(), bins=30, kde=True, color='teal')
plt.title('Distribusi Tahun Lahir Idol K-pop')
plt.xlabel('Tahun Lahir')
plt.ylabel('Frekuensi')
plt.show()
```

**Insight:**
- Terdapat tren peningkatan jumlah kelahiran idol yang masuk industri di rentang tahun 1995-2005.
- Puncak distribusi menunjukkan bahwa sebagian besar idol aktif merupakan generasi Z.


## 4. Kesimpulan Akhir
- **Karakteristik Fisik**: Terdapat standar fisik yang cukup jelas antara gender di industri ini.
- **Internasionalisasi**: Meskipun tetap berpusat di Korea, talenta dari mancanegara semakin beragam.
- **Keseimbangan Industri**: Keseimbangan jumlah idol laki-laki dan perempuan menunjukkan pasar yang stabil untuk keduanya.
- **Modernisasi**: Tren menunjukkan dominasi idol muda yang lahir di era digital.
